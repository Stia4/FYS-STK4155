# Analysis on data

In the following cells we take a small 30 $\times$ 30 section from the given data "SRTM_data_Norway_1.tif" and perform the same analysis as we have up until now on this data. We find the MSE using OLS, perform Boostrap on the data, use k-folds and use RIDGE and LASSO regression.  

```python
""" Load data """
terrain1 = imread('data/SRTM_data_Norway_1.tif')
x_pos = 1000
y_pos = 2340
dl = 30
z = terrain1[y_pos:y_pos+dl,x_pos:x_pos+dl]
z = z - np.mean(z) # normalize by mean

""" Plot data """
fig, ax = plt.subplots()
ax.imshow(terrain1)

axins = ax.inset_axes([1.25, 0.25, 1, 0.5])
terrain1_tmp = np.zeros(np.shape(terrain1)) + 72 # <- raise zeros to 72 to get correct coloring ...
terrain1_tmp[y_pos:y_pos+dl,x_pos:x_pos+dl] = terrain1[y_pos:y_pos+dl,x_pos:x_pos+dl]
axins.imshow(terrain1_tmp, extent=[0,1801,3601,0])
axins.set_xlim(x_pos, x_pos+dl-0.5)
axins.set_ylim(y_pos, y_pos+dl-0.5)
axins.set_xticklabels([0, 10, 20, 30])
axins.set_yticklabels([0, 5, 10, 15, 20, 25, 30])
rec, lin = ax.indicate_inset_zoom(axins, edgecolor="red", linewidth=1)
[lini.set(lw=2) for lini in lin]

plt.tight_layout()
plt.savefig("data.pdf")
plt.show()

print("Shape terrain:", np.shape(terrain1), ", Shape z:", np.shape(z))
print("Standard deviation:", np.std(z), ", Max height difference:", np.max(z) - np.min(z))
```

```python
x_ = np.linspace(0, 1, len(z))
y_ = np.linspace(0, 1, len(z[1]))
x, y = np.meshgrid(x_, y_)

test_size = 4
orders = np.arange(0,15)

from Project1b import MSE, train_test_data_evenly, fit_OLS,R2_Score

x_lrn, y_lrn, z_lrn, x_tst, y_tst, z_tst = train_test_data_evenly(x, y, z, test_size)
X_lrn, X_tst, beta_OLS, z_lrn_model, z_tst_model = fit_OLS(orders, x_lrn, y_lrn, z_lrn, x_tst, y_tst, z_tst)

MSE_lrn = []
R2_lrn  = []
MSE_tst = []
R2_tst  = []
for i in orders:
    MSE_lrn.append(MSE(z_lrn, z_lrn_model[i]))
    R2_lrn.append(R2_Score(z_lrn, z_lrn_model[i]))
    MSE_tst.append(MSE(z_tst, z_tst_model[i]))
    R2_tst.append(R2_Score(z_tst, z_tst_model[i]))

plt.plot(orders, MSE_lrn, "o-", label="MSE Training data")
plt.plot(orders, MSE_tst, "o-", label="MSE Test data")
plt.hlines(0, min(orders), max(orders), colors="gray", linestyles="dashed")
plt.xticks(ticks=orders)
plt.xlabel("Polynome order")
plt.ylabel("Mean squared error")
plt.legend(loc=1)
plt.grid(alpha = 0.3)
plt.tight_layout()
plt.yscale("log")
plt.savefig("MSE_data.pdf", format="pdf")
plt.show()

plt.plot(orders, R2_lrn, "o-", label="R2 Score Training data")
plt.plot(orders, R2_tst, "o-", label="R2 Score Test data")
plt.hlines(1, min(orders), max(orders), colors="gray", linestyles="dashed")
plt.xticks(ticks=orders)
plt.xlabel("Polynome order")
plt.ylabel("R2 Score")
plt.legend(loc=4)
plt.tight_layout()
plt.grid(alpha = 0.3)
plt.savefig("R2_data.pdf", format="pdf")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.utils import resample

from Project1c import Bootstrap
n_bootstraps = 90 
maxdegree = 15
x_ = np.append(x.ravel().reshape(-1, 1), y.ravel().reshape(-1, 1), axis=1)
y_ = z.ravel()
y_ = y_.reshape(y_.size, 1)
bias, variance, error, degrees = Bootstrap(x_, y_, n_bootstraps, maxdegree, silent=True)

plt.plot(degrees, error, label='Error')
plt.plot(degrees, bias, label='bias')
plt.plot(degrees, variance, label='Variance')
plt.legend()
plt.xlabel("Polynome order")
plt.grid(alpha = 0.3)
plt.yscale('log')
plt.savefig("c_data.pdf")
plt.show()
```

```python
from Project1d import kfold
### Setting parameters
k = 50
orders = np.arange(15)

### Generating data and splitting into k folds of learn/test data
lrn, tst = kfold(x, y, z, k, seed=0)

### Iterating over folds
running_MSE = np.zeros_like(orders, dtype=float)
for i in range(k):
    ## Fitting polynomials of all orders to fold i learn set, then applying to test data
    z_tst_model = fit_OLS(orders, lrn[0, i], lrn[1, i], lrn[2, i],
                                  tst[0, i], tst[1, i], tst[2, i])[-1]
    for j in orders:
        running_MSE[j] += MSE(tst[2, i], z_tst_model[j]) / k

plt.plot(orders, running_MSE)
plt.hlines(0, min(orders), max(orders), colors="gray", linestyles="dashed")
plt.xticks(ticks=orders)
plt.xlabel("Polynome order")
plt.ylabel("Mean squared error")
plt.title(f"MSE for k-fold cross validation, k = {k}")
plt.grid(alpha = 0.3)
plt.yscale('log')
plt.tight_layout()
plt.savefig("d_data.pdf")
plt.show()
```

```python
from Project1e import Bootstrap_pred
from tqdm import trange
### Defining input parameters
n_degrees = 20                     # number of degrees, degrees 0 ... n_degrees-1
nk = 4                             # number of k-folds
n_bootstraps = 15                  # number of times to resample each learn-set
lambdas = np.logspace(-15, -5, 20) # values of lambda to use in Ridge regression

terrain1 = imread('data/SRTM_data_Norway_1.tif')
z = terrain1[y_pos:y_pos+dl,x_pos:x_pos+dl]
z = z - np.mean(z)

x_ = np.linspace(0, 1, len(z))
y_ = np.linspace(0, 1, len(z[1]))
x, y = np.meshgrid(x_, y_)



### Making dataset
learn_set, test_set = kfold(x, y, z, nk, seed=0)
bias_ = []
variance_ = []
error_ = []

### Looping over lambdas, each generating a set of bias/variance/error curves
for i, lam in tqdm(enumerate(lambdas)):
    bias     = np.zeros(n_degrees)
    variance = np.zeros(n_degrees)
    error    = np.zeros(n_degrees)

    ### Looping over k-folds, using a different combination each time
    for k in range(nk):
        x_lk, y_lk, z_lk = learn_set[:, k] # x,y,z for learn set nr. k
        x_tk, y_tk, z_tk =  test_set[:, k] # x,y,z for test set

        # Performing n_bootstraps resamplings of learn set and resulting n_degrees models to each test set
        z_pred = Bootstrap_pred(x_lk, x_tk, y_lk, y_tk, z_lk, z_tk, n_bootstraps, n_degrees, lam) # <- Note lambda is input here

        # Finding how well each model fits across all bootstraps, doing running average over all k-folds
        for degree in range(n_degrees): #         Broadcasting ↓   Axis with bootstrapped values ↓
            error[degree]    += np.mean( np.mean((z_tk[:, np.newaxis] - z_pred[degree])**2, axis=1, keepdims=True) ) / nk # <- Note division by nk to
            bias[degree]     += np.mean( (z_tk[:, np.newaxis] - np.mean(z_pred[degree], axis=1, keepdims=True))**2 ) / nk #    get a running average
            variance[degree] += np.mean( np.var(z_pred[degree], axis=1, keepdims=True) ) / nk
            
    bias_.append(bias)
    variance_.append(variance)
    error_.append(error)
```

```python
error_ = np.array(error_)

degrees = np.arange(n_degrees)
extent = [degrees[0], degrees[-1], np.log10(lambdas[0]), np.log10(lambdas[-1])]

plt.rcParams["figure.figsize"] = (12, 8)
plt.rcParams.update({'font.size': 22})
plt.contourf(np.log10(error_), extent=extent, levels=30)
ix, iy = np.where(np.min(error_) == error_)
plt.plot(degrees[iy], np.log10(lambdas[ix]), 'rx', label=r'pol. degree = %.i, $\lambda$ = %.2e' % (degrees[iy], lambdas[ix]))
plt.xticks(degrees[::2])
#plt.yticks(np.log10(lambdas))
plt.xlabel("Polynomial degree")
plt.ylabel(r"log10($\lambda$)")
plt.legend()
plt.colorbar(label="log10(MSE)")
np.random.seed(0)
plt.tight_layout()
plt.savefig("RIDGE_data.pdf")
```
